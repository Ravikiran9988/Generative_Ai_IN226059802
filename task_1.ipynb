Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" are treated as different words if case normalization is not applied. This can increase the vocabulary size unnecessarily and create inconsistency. To avoid this, we usually convert all text to lowercase so both are treated as the same word.

### 2. What happens if stopwords are not removed?
If stopwords like "is", "the", and "and" are not removed, the dataset becomes noisy. These words appear very frequently but don’t add much meaning, which can reduce model efficiency and increase processing time.

### 3. Real-world scenarios where removing stopwords can be harmful
- In sentiment analysis, removing words like "not" can completely change the meaning (e.g., "not good" becomes "good").
- In question-answering systems, words like "what", "who", and "where" are important for understanding the query.

### 4. Difference between stemming and lemmatization
Stemming reduces words to their root form by removing suffixes, but it may not always produce meaningful words (e.g., "running" → "run").  
Lemmatization converts words to their correct base form using language rules (e.g., "better" → "good"), making it more accurate.


Task 2: Preprocessing Function

```python
import re

def preprocess_text(text):
    # Handle invalid input
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs and emails
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove emojis and special characters
    text = re.sub(r'[^\w\s]', '', text)

    # Handle repeated characters (e.g., "soooo" → "so")
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words (≤2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    results.append((sentence, tokens, cleaned))

for r in results:
    print("\nOriginal Text:", r[0])
    print("Cleaned Tokens:", r[1])
    print("Cleaned Sentence:", r[2])
```

Task 4: Token Analytics

```python
import numpy as np

analytics = []

for sentence, tokens, cleaned in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(t) for t in tokens]) if tokens else 0

    analytics.append({
        "sentence": sentence,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_token_length": avg_length
    })

for a in analytics:
    print("\nSentence:", a["sentence"])
    print("Total Tokens:", a["total_tokens"])
    print("Unique Tokens:", a["unique_tokens"])
    print("Avg Token Length:", round(a["avg_token_length"], 2))
```

### Observations

- The sentence "Win $$$ now!!! Limited offer!!!" had the most noise because it contained symbols, repeated punctuation, and promotional content.

- The sentence "I absolutely looooved this product 😍😍" retained the most meaningful tokens after cleaning, as it clearly expresses strong sentiment even after preprocessing.

Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:", counter.most_common(10))
print("Top 5 Least Frequent Words:", counter.most_common()[-5:])
```

### Explanation

The most frequent words are common terms that appear across multiple sentences.  
The least frequent words are usually more specific and appear only once.  

This helps in understanding word distribution in the dataset.

Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

Task 7: Error Handling

```python
edge_cases = [
    "",
    "😂😂😂😂😂",
    "1234567890"
]

for case in edge_cases:
    tokens, cleaned = preprocess_text(case)
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
```

### Edge Case Observations

- Empty input returns no tokens and an empty sentence.
- Emoji-only input is cleaned completely, resulting in no tokens.
- Number-only input removes all numbers, leaving no meaningful text.

This shows that the preprocessing function handles edge cases properly without errors.
