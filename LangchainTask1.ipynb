Task 4: ChatPromptTemplate System

```python
from langchain_core.prompts import ChatPromptTemplate

def get_chat_prompt(role, topic):

    if role == "teacher":
        system_msg = "You are a helpful teacher who explains concepts clearly."
    elif role == "interviewer":
        system_msg = "You are an interviewer asking insightful questions."
    elif role == "motivator":
        system_msg = "You are a motivational speaker who inspires learners."
    else:
        system_msg = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Talk about {topic}")
    ])

    return chat_template.format_messages(topic=topic)

# Example
messages = get_chat_prompt("teacher", "Neural Networks")
print(messages)
```

Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        raise ValueError("Invalid audience! Choose from beginner, intermediate, expert")

    if tone not in valid_tone:
        raise ValueError("Invalid tone! Choose from formal, casual, fun")

    return True

# Example
validate_inputs("beginner", "fun")
```

Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):

    validate_inputs(audience, tone)

    templates = {
        "teaching": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        ),

        "interview": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask interview questions about {topic} for {audience} in a {tone} tone"
        ),

        "storytelling": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )
    }

    if style not in templates:
        raise ValueError("Invalid style! Choose teaching/interview/storytelling")

    return templates[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

Task 7: Template Reusability Test

```python
reuse_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}"
)

inputs = [
    {"topic": "AI", "audience": "beginner"},
    {"topic": "Python", "audience": "kids"},
    {"topic": "Data Science", "audience": "students"},
    {"topic": "Blockchain", "audience": "developers"},
    {"topic": "Cybersecurity", "audience": "professionals"}
]

for inp in inputs:
    print(reuse_template.format(**inp))
```
